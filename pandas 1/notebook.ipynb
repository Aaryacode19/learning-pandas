Summary Statistics 

```python
#finding mean mode and median of our new column
mean = file["New column"].mean()
print(f"mean is: {mean}")
mode = file["New column"].mode()
print(f"mode is: {mode}")
median = file["New column"].median()
print(f"median is: {median}")
```

```python
#finding std variance minimum value maximum value and range
max = file["New column"].max()
print(f"maxm no in our data is: {max}")
min = file["New column"].min()
print(f"minm no in our data is: {min}")
variance = file["New column"].var()
print(f"variance of data is {variance}")
std = file["New column"].std()
print(f"std of data is: {std}")
range = max - min
print(f"range is: {range}")
```

```python
"""
The .agg() method allows you to apply your own custom functions to a DataFrame, as well as apply functions to more than one column of a DataFrame at once, making your aggregations super-efficient.
"""
x = file["New column"].agg(['mean', 'median'])
print(x)
#so what agg() did here is after we accessed a specific column agg() helped us to find mean and median both of that column at same time in less lines of code.
```

```python
#using Cumulative statistics
"""
- **Cumulative Maximum**: The maximum value encountered so far at each point.
- **Cumulative Minimum**: The minimum value encountered so far at each point.
- **Cumulative Sum**: The sum of all values up to each point.
- **Cumulative Product**: The product of all values up to each point.
"""
cumulative_sum = file["New column"].cumsum()
print(f"cumsum is: {cumulative_sum}")
cumuative_min = file["New column"].cummin()
print(f"cumin is: {cumuative_min}")
cumulative_max = file["New column"].cummax()
print(f"cummax is : {cumulative_max}")
cumulative_prod = file["New column"].cumprod()
print(f"cumprod is : {cumulative_prod}")
```

```python
#dropping duplicates from our dataset
drop_duplicate_departments = file.drop_duplicates(subset="Department") #note: here subset means column name which contains duplicate values.
print(f"our sorted data after dropping duplicate values is: {drop_duplicate_departments}")
#so here this method removed all those rows from our dataset which  has duplicate department. now each particapation is of different department so our data is sorted on basis of drop duplicate dpts.
#similarly u can take more subsets/ columns to sort ur data
drop_multiple_duplicates = file.drop_duplicates(subset=["Department", "Time of Registration"])
drop_multiple_duplicates
#so here we sorted data on basis of both duplicate depts n time
```

```python
# Counting categorical variables using value_counts()

"""
The value_counts() function in pandas is used to count the number of occurrences of each unique value in a Series or a DataFrame column. 
It returns a Series containing counts of unique values. This is particularly useful for categorical data to understand the distribution of categories.
"""

# Example: Counting occurrences of each department in the "Department" column
department_counts = file["Department"].value_counts()

# Display the counts of each department
department_counts

# You can also use value_counts() with other columns, for example:
# Counting occurrences of each city in the "City" column
city_counts = file["City"].value_counts()

# Display the counts of each city
city_counts
```

```python
# Grouped Summary Statistics using groupby() and describe()

"""
Grouped summary statistics involve calculating summary statistics (like mean, median, count, etc.) for different groups within a dataset. 
In pandas, this can be achieved using the groupby() function followed by the describe() function to get a comprehensive summary of statistics for each group.
"""

# Example: Grouping by "Department" and calculating summary statistics for the "Year of Study" column
grouped_summary = file.groupby("Department")["Year of Study"].describe()

# Display the grouped summary statistics
grouped_summary

# You can also group by multiple columns and calculate summary statistics for multiple columns
# Example: Grouping by "Department" and "City" and calculating summary statistics for the "Year of Study" and "Shirt Size" columns
grouped_summary_multiple = file.groupby(["Department", "City"])[["Year of Study", "Shirt Size"]].describe()

# Display the grouped summary statistics for multiple columns
grouped_summary_multiple
```

Pivot Tables

```python
file.pivot_table(values="Department", index="Year of Study", aggfunc='count')
#Here values argument is the column u wanna summarise nd index column in the column u wanna group by, aggfuncn is the funcn u wanna perform on selected columns. like here we got the no of count of depts in 1st 2nd so on yrs.
```

```python
import pandas as pd
import numpy as np

# multiple argumnts
file.pivot_table(values="New column", index="Year of Study", aggfunc=[np.mean, np.median])
```

```python
#pivot on two values
file.pivot_table(values='Department', index='Year of Study', columns='Registration ID', aggfunc='first')

#here columns is the arg for 2nd value after values arg wecan pass more values in columns arg
```

```python
#handling missing values
file.pivot_table(values='Department', index='Year of Study', columns='Registration ID', aggfunc='first', fill_value = 0)
```

```python
# summing with pivot tables
file.pivot_table(values='Department', index='Year of Study', columns='Registration ID', aggfunc='count', fill_value=0, margins=True)

#using margins = true we can see multiple summary statistics
```

## Handling missing values

```python
# Create a DataFrame with missing values
import numpy as np
miss_data = {
    'Date': pd.date_range(start='2024-01-01', periods=5, freq='D'),
    'Sales': [200, np.nan, 150, np.nan, 300],
    'Revenue': [np.nan, 5000, 4000, 4500, np.nan],
    'Category': ['A', 'B', 'A', 'B', 'A']
}

missing = pd.DataFrame(miss_data)

missing
```

```python
#detecting missing values 
missing.isna()
```

```python
#detecting missing values
missing.isnull().sum()
```

```python
#Plotting missing values
import matplotlib.pyplot as plt 
missing.isna().sum().plot(kind="bar") 
plt.show()
```

```python
#dropping or Removing missing values
missing.dropna() 
```

```python
#replacing or filling missing values using 0
missing.fillna(0) 
```
